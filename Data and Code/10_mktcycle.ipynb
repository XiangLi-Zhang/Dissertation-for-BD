```python
## 导入NumPy、Pandas、Wind
import numpy as np
import pandas as pd
from scipy import stats
from WindPy import w
import datetime
```

```python
## 启动Wind终端的API
## 判断WindPy是否已经登录成功
w.start() 
w.isconnected()
```

```python
## 指数每日收盘数据
mkt_p = w.wsd("000002.SH", "close", "2006-10-01", "2019-03-31", "")
mkt_p = pd.DataFrame(mkt_p.Data, index = mkt_p.Codes, columns = mkt_p.Times)
mkt_p = mkt_p.T
mkt_p = mkt_p.reset_index()
mkt_p.columns = ['date', 'price']
mkt_p['date'] = pd.to_datetime(mkt_p['date'])
mkt_p['year'] = mkt_p['date'].dt.year
mkt_p['month'] = mkt_p['date'].dt.month
```

```python
mkt_mp = mkt_p.groupby([mkt_p['year'], mkt_p['month']])['price'].agg([('num', 'count'), ('meanprice', 'mean'),])
mkt_mp = mkt_mp.reset_index()
```

```python
mktmp_03 = mkt_mp.head(144).reset_index(drop = True)
mktmp_02 = mkt_mp.drop(0, axis = 0).head(144).reset_index(drop = True)
mktmp_01 = mkt_mp.drop([0, 1], axis = 0).head(144).reset_index(drop = True)
mktmp_0 = mkt_mp.drop([0, 1, 2], axis = 0).head(144).reset_index(drop = True)
mktmp_1 = mkt_mp.drop([0, 1, 2, 3], axis = 0).head(144).reset_index(drop = True)
mktmp_2 = mkt_mp.drop([0, 1, 2, 3, 4], axis = 0).head(144).reset_index(drop = True)
mktmp_3 = mkt_mp.drop([0, 1, 2, 3, 4, 5], axis = 0).head(144).reset_index(drop = True)
```

```python
mkt_cycle = mktmp_0[['year', 'month', 'meanprice']]
mkt_cycle['cp03'] = mktmp_0['meanprice'] > mktmp_03['meanprice']
mkt_cycle['cp02'] = mktmp_0['meanprice'] > mktmp_02['meanprice']
mkt_cycle['cp01'] = mktmp_0['meanprice'] > mktmp_01['meanprice']
mkt_cycle['cp1'] = mktmp_0['meanprice'] > mktmp_1['meanprice']
mkt_cycle['cp2'] = mktmp_0['meanprice'] > mktmp_2['meanprice']
mkt_cycle['cp3'] = mktmp_0['meanprice'] > mktmp_3['meanprice']
```

```python
mkt_cycle['cc'] = mkt_cycle['cp03'] + 0
mkt_cycle['cc'] = mkt_cycle['cp02'] + 0 + mkt_cycle['cc']
mkt_cycle['cc'] = mkt_cycle['cp01'] + 0 + mkt_cycle['cc']
mkt_cycle['cc'] = mkt_cycle['cp1'] + 0 + mkt_cycle['cc']
mkt_cycle['cc'] = mkt_cycle['cp2'] + 0 + mkt_cycle['cc']
mkt_cycle['cc'] = mkt_cycle['cp3'] + 0 + mkt_cycle['cc']
```

```python
mkt_cycle.sort_values(by= ['cc', 'year', 'month'] , ascending = True).to_csv("Rgrss/mkt_cycle.csv", index = False)
```
