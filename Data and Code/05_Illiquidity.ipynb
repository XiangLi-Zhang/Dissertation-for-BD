```python
vlm0 = pd.read_csv('Raw Transaction Data/vlm')
```

```python
vlm_d1  = pd.melt(vlm0, 
            id_vars = 'index', 
            value_vars =list(vlm0 .columns[1:]), 
            var_name='code', 
            value_name='volumn')
vlm_d1.columns = ['date', 'code', 'volumn']
vlm_d1['date'] = pd.to_datetime(vlm_d1['date'])
vlm_d1['year'] = vlm_d1['date'].dt.year
vlm_d1['month'] = vlm_d1['date'].dt.month
vlm_d1['volumn'] = vlm_d1['volumn'] / 1000000
```

```python
## 合并pctchg与vlm数据
ilqdt_dt = pd.merge(pctdt_d1, vlm_d1, on = ['code', 'date'], how = 'left')
ilqdt_dt = ilqdt_dt.dropna(subset = ['pctchg', 'volumn'])
ilqdt_dt = ilqdt_dt[['code', 'date', 'year', 'month', 'pctchg', 'volumn']]
ilqdt_dt = ilqdt_dt[ilqdt_dt['volumn'] != 0]
ilqdt_dt['indct'] = ilqdt_dt['pctchg'] / ilqdt_dt['volumn'] 
```

```python
ilqdt_dt
```

```python
## 计算各公司非流动性
illiq_cmp = ilqdt_dt.groupby([ilqdt_dt['code'], ilqdt_dt['year'], ilqdt_dt['month']])['indct'].mean()
illiq_cmp = illiq_cmp.reset_index()
```

## 等权重加权法

```python
illiq_ew = illiq_cmp.groupby([illiq_cmp['year'], illiq_cmp['month']])['indct'].agg([('illiq', 'mean'),])
illiq_ew = illiq_ew.reset_index()
```

```python
illiq_ew1 = illiq_ew.head(97).drop(index = 0, axis = 0).reset_index(drop = True)
illiq_ew1["illiq"] = np.log(illiq_ew1["illiq"])
illiq_ew0 = illiq_ew.head(96)
illiq_ew0["illiq"] = np.log(illiq_ew0["illiq"])
```

```python
illiq_ewx = sm.add_constant(illiq_ew0["illiq"])
model_ew = sm.OLS(illiq_ew1["illiq"], illiq_ewx)
results_ew = model_ew.fit()
print(results_ew.summary())
```

```python
ILLIQE_ew = -2.0594 + 0.6944 * illiq_ew['illiq']
ILLIQE_ew.head(96).to_csv("Rgrss/ILLIQE_ew.csv", index = False)
```

## 市值加权法

```python
## 合并权重数据，加权
illiq = pd.merge(illiq_cmp, weight, on = ['code', 'year', 'month'], how = 'left')
illiq['illiq'] = illiq['indct'] * illiq['weight']
```

```python
## 计算市场不流动性
illiq_mkt = illiq.groupby([illiq['year'], illiq['month']])['illiq'].agg([('Amihud', 'sum'),])
illiq_mkt = illiq_mkt.reset_index()
```

```python
illiq_mkt.to_csv("Rgrss/illiq_vw.csv", index = False)
```
