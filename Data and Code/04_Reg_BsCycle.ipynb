# 剔除经济周期波动的影响(宏观数据，与板块无关)

```python
## 导入NumPy、Pandas、Wind
import numpy as np
import pandas as pd
from scipy import stats
from WindPy import w
import datetime
```

```python
w.start() # 默认命令超时时间为120秒，如需设置超时时间可以加入waitTime参数，例如waitTime=60,即设置命令超时时间为60秒  
w.isconnected() # 判断WindPy是否已经登录成功
```

```python
## 读取宏观经济数据
cyc_data = w.edb("M0001382,M0001384,M0043814,M0039354,M0024063", "2007-01-01", "2018-12-31")
cyc = pd.DataFrame(cyc_data.Data,index = cyc_data.Codes,columns = cyc_data.Times)
cyc = cyc.T
cyc = cyc.reset_index()
cyc.columns = ['date', 'M1', 'M2', 'LR', 'GDPgrow', 'FE']
```

```python
## tidy之后的cyc
tcyc = cyc.fillna(method = 'bfill')
tcyc['DM'] = (tcyc['M2'] - tcyc['M1'])/tcyc['M1']
tcyc['LR'] = tcyc['LR'] / 100
tcyc['GDPGR'] = tcyc['GDPgrow'] /100
tcyc['FE'] = tcyc['FE'] / 100
```

```python
## 保留宏观经济数据并保存
fmacro = tcyc[['DM', 'LR', 'FE', 'GDPGR']]
```

```python
fmacro.to_csv("Rgrss/fmacro.csv", index = False)
```
