# 绘制折线图等描述性图片

```python
## 导入NumPy、Pandas
import numpy as np
import pandas as pd
import datetime
from plotnine import *
```

```python
## 读取数据
## variables = pd.read_excel("Rgrss/variables.xlsx")
variables = pd.read_csv("Plot/reg_data.csv")
variables['month'] = variables.index % 12 + 1
variables['day'] = 1
variables['date'] = pd.to_datetime(variables[['year', 'month', 'day']])
variables = variables.drop(['year', 'month', 'day'],axis=1)
```

## 方差对比

```python
## 合并市场方差、等权重加权方差、市值加权方差
var_value = pd.melt(variables[['date', 'mkt_var', 'var_ew', 'var_vw']], 
    id_vars = 'date', var_name = 'VarType', value_name = 'VarValue')
var_value['VarType'] = var_value['VarType'].astype('category').cat.rename_categories(["市场方差", "等权重加权方差", "市值加权方差"])
```

```python
var_value = variables[['date', 'mkt_var', 'var_ew', 'var_vw']]
```

```python
var_plot1 = (ggplot(var_value)
+ theme_classic() 
+ geom_line(mapping = aes(x = 'date', y = 'mkt_var'), stat = 'identity', colour = "red")
+ labs(x = "年份",y = "市场方差") 
+ scale_x_datetime(breaks = "1 year", date_labels = '%Y')
+ scale_y_continuous(limits = [0, 0.3])
+ theme(legend_position = "none", # legend_title = element_blank(),
        text = element_text(family = "Microsoft YaHei") )
)

var_plot2 = (ggplot(var_value)
+ theme_classic() 
+ geom_line(mapping = aes(x = 'date', y = 'var_ew'), stat = 'identity', colour = "green")
+ labs(x = "年份",y = "等权重加权方差") 
+ scale_x_datetime(breaks = "1 year", date_labels = '%Y')
+ scale_y_continuous(limits = [0, 0.3])
+ theme(legend_position = "none", # legend_title = element_blank(),
        text = element_text(family = "Microsoft YaHei") )
)

var_plot3 = (ggplot(var_value)
+ theme_classic() 
+ geom_line(mapping = aes(x = 'date', y = 'var_vw'), stat = 'identity', colour = "blue")
+ labs(x = "年份",y = "市值加权方差") 
+ scale_x_datetime(breaks = "1 year", date_labels = '%Y')
+ scale_y_continuous(limits = [0, 0.3])
+ theme(legend_position = "none", # legend_title = element_blank(),
        text = element_text(family = "Microsoft YaHei") )
)
```

```python
var_plot1.save(filename = "Plot/var_plot1.png", width = 400, height=50, units='mm')
var_plot2.save(filename = "Plot/var_plot2.png", width = 400, height=50, units='mm')
var_plot3.save(filename = "Plot/var_plot3.png", width = 400, height=50, units='mm')
```
