# 进行样本外检验

```python
## 导入NumPy、Pandas
import numpy as np
import pandas as pd
import datetime
import statsmodels.api as sm
```

```python
## 读取r0数据
r0 = pd.read_csv("Rgrss/r0.csv")
## 读取计算好的各项数据
regdt = pd.read_csv("Plot/reg_data.csv")
```

```python
Ave_Rm = r0.head(96).mean()
reg_tr = regdt.head(96)
```

## 市场指标的样本外方差

```python
## 使用市场方差作为预测因子
X_mktvar = sm.add_constant(reg_tr["mkt_var"])
model_mktv = sm.OLS(reg_tr["r"], X_mktvar)
results_mktv = model_mktv.fit()

## 根据回归系数计算预期市场收益，并计算oos等数据
DF_mv = pd.DataFrame({
    'r_lead': regdt['r'],
    'mu':results_mktv.params[0] + results_mktv.params[1] * regdt["mkt_var"],
    'r_mean': Ave_Rm[0],
    })
OOS_mv = DF_mv[DF_mv.index > 95]
OOS_mv = OOS_mv.reset_index(drop = True)
OOS_mv['del1'] = (OOS_mv['r_lead'] - OOS_mv['mu'])**2
OOS_mv['del2'] = (OOS_mv['r_lead'] - OOS_mv['r_mean'])**2
MSEp_mv = OOS_mv['del1'].sum() / len(OOS_mv)
MSEn_mv = OOS_mv['del2'].sum() / len(OOS_mv)
Roos_mv = 1 - MSEp_mv / MSEn_mv
AveRoos_mv = Roos_mv - (1/(len(OOS_mv) - 1 - 1)) * (1 - Roos_mv)

## 计算ENC
OOS_mv['enc'] = (OOS_mv['r_lead'] - OOS_mv['r_mean'])**2 - (OOS_mv['r_lead'] - OOS_mv['r_mean']) * (OOS_mv['r_lead'] - OOS_mv['mu'])
ENC_mv = (len(OOS_mv) - 1 + 1) / len(OOS_mv) * OOS_mv['enc'].sum() / MSEp_mv

## 打印各项计算结果
print("MSEp：%.3f MSEn：%.3f Roos: %.3f AveRoos: %.3f ENC: %.3f" %(MSEp_mv, MSEn_mv, Roos_mv, AveRoos_mv, ENC_mv))
```

```python
## 使用市场偏度作为预测因子
X_ms = sm.add_constant(reg_tr["mkt_skew"])
model_ms = sm.OLS(reg_tr["r"], X_ms)
results_ms = model_ms.fit()

## 根据回归系数计算预期市场收益，并计算oos等数据
DF_ms = pd.DataFrame({
    'r_lead': regdt['r'],
    'mu': results_ms.params[0] + results_ms.params[1] * regdt["mkt_skew"],
    'r_mean': Ave_Rm[0],
    })
OOS_ms = DF_ms[DF_ms.index > 95]
OOS_ms = OOS_ms.reset_index(drop = True)
OOS_ms['del1'] = (OOS_ms['r_lead'] - OOS_ms['mu'])**2
OOS_ms['del2'] = (OOS_ms['r_lead'] - OOS_ms['r_mean'])**2
MSEp_ms = OOS_ms['del1'].sum() / len(OOS_ms)
MSEn_ms = OOS_ms['del2'].sum() / len(OOS_ms)
Roos_ms = 1 - MSEp_ms / MSEn_ms
AveRoos_ms = Roos_ms - (1/(len(OOS_ms) - 1 - 1)) * (1 - Roos_ms)

## 计算ENC
OOS_ms['enc'] = (OOS_ms['r_lead'] - OOS_ms['r_mean'])**2 - (OOS_ms['r_lead'] - OOS_ms['r_mean']) * (OOS_ms['r_lead'] - OOS_ms['mu'])
ENC_ms = (len(OOS_ms) - 1 + 1) / len(OOS_ms) * OOS_ms['enc'].sum() / MSEp_ms

## 打印各项计算结果
print("MSEp：%.3f MSEn：%.3f Roos: %.3f AveRoos: %.3f ENC: %.3f" %(MSEp_ms, MSEn_ms, Roos_ms, AveRoos_ms, ENC_ms))
```
